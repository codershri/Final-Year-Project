```python
import pydicom
import os
import numpy as np
from PIL import Image
```

```python
# def dicom_to_jpg(dicom_path, jpg_path):
#     # Read DICOM file
#     dicom_image = pydicom.dcmread("D:\\Dicom\\case1")
    
#     # Get pixel array from DICOM file
#     image_array = dicom_image.pixel_array

#     # Normalize pixel values to the range [0, 255]
#     image_array = (image_array - np.min(image_array)) / (np.max(image_array) - np.min(image_array)) * 255.0
#     image_array = image_array.astype(np.uint8)

#     # Save as JPG
#     image = Image.fromarray(image_array)
#     image.save("D:\\Dicom\\case1_jpg")

# # Example usage
# dicom_to_jpg("path_to_dicom.dcm", "output_image.jpg")

def dicom_to_jpg(dicom_path, jpg_path):
    # Read DICOM file
    dicom_image = pydicom.dcmread(dicom_path)
    
    # Get pixel array from DICOM file
    image_array = dicom_image.pixel_array

    # Normalize pixel values to the range [0, 255]
    image_array = (image_array - np.min(image_array)) / (np.max(image_array) - np.min(image_array)) * 255.0
    image_array = image_array.astype(np.uint8)

    # Save as JPG
    image = Image.fromarray(image_array)
    image.save(jpg_path)

def convert_all_dicoms_in_folder(dicom_folder, output_folder):
    # Ensure output folder exists
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)
    
    # Loop through all files in the dicom folder
    for filename in os.listdir(dicom_folder):
        if filename.endswith(".dcm"):  # Filter DICOM files
            dicom_path = os.path.join(dicom_folder, filename)
            jpg_filename = filename.replace(".dcm", ".jpg")  # Change extension to JPG
            jpg_path = os.path.join(output_folder, jpg_filename)
            
            # Convert DICOM to JPG
            dicom_to_jpg(dicom_path, jpg_path)
            print(f"Converted {filename} to {jpg_filename}")

# Example usage
dicom_folder = "D:\\Dicom\\case1"  # Folder where DICOM files are located
output_folder = "D:\\Dicom\\case1_jpg"  # Folder to save converted JPGs

convert_all_dicoms_in_folder(dicom_folder, output_folder)



```
